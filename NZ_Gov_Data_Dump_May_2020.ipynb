```
# THIS IS THE PRODUCTION VERSION
from urllib import request
from bs4 import BeautifulSoup
import copy
import wget

# CHECK YOUR PARSER LOCATION ON NEXT LINE
release_url = "https://covid19.govt.nz/resources/key-documents-and-legislation/proactive-release/supporting-the-economy/"
page_data = request.urlopen(release_url)
the_page = page_data.read()

chowder = BeautifulSoup(the_page, "html.parser")

bisque_documents = chowder.find("div", class_="content-element__content")

bisque_copy = copy.copy(bisque_documents)

bisque2_documents = bisque_copy.find_all("a") 

download_directory = "downloads/supporting_the_economy/" # CHECK YOUR DOWNLOAD DIRECTORY LOCATION. THIS SHOULD MATCH WHATEVER DIRECTORY(IES) YOU CREATED EARLIER.

for current_line in bisque2_documents:
  temp_line = str(current_line)
  line_length = len(temp_line)
  first_index = '"'
  second_index = '">'
  first_quote_index = temp_line.find(first_index)
  second_quote_index = temp_line.find(second_index)
  extracted_link = temp_line[(first_quote_index+1):second_quote_index]
  start_of_link = "https://covid19.govt.nz"
  combined_link = start_of_link + extracted_link
  print ("DOWNLOADING", combined_link)
  wget.download(combined_link, out=download_directory)

```
